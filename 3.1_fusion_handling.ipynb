```python
import pandas as pd
import numpy as np
import pickle
from pandas_profiling import ProfileReport
from missingno import matrix
import pyreadstat
```

```python
df_ger = pd.read_pickle("data/ger_preimp.p")
df_usa = pd.read_pickle("data/usa_preimp.p")

df_ger['country'] = 0 #Germany
df_ger['recipient_status'] = df_ger['recipient_status'].replace({3: 2, 4: 2})

df_usa['country'] = 1 #USA
df_usa.rename(columns={'graft_1yr': '1yr_graft'}, inplace=True)
```

```python
common_columns = df_ger.columns.intersection(df_usa.columns)
print(common_columns)
```

```python
len(common_columns)
```

```python
merged_df = pd.concat([df_ger[common_columns], df_usa[common_columns]], ignore_index=True)
merged_df
```

```python
merged_df = merged_df.round(2) 
```
